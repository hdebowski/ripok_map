```python
#!pip install geemap
import folium
import pandas as pd
import ee
import geemap.foliumap as geemap
```

```python
cities = pd.read_excel('dane_oczyszczone.xlsx')
```

```python
cities.head(38)
```

```python
 def select_marker_color(row):
  if row['Rodzaj instalacji'] == 'Regionalne instalacje do mechaniczno-biologicznego przetwarzania zmieszanych odpadów komunalnych':
    return 'beige'
  elif row['Rodzaj instalacji'] == 'Regionalne instalacje do biologicznego przetwarzania odpadów zielonych i innych bioodpadów':
    return 'green'
  elif row['Rodzaj instalacji'] == 'Regionalne instalacje do składowania odpadów powstających w procesie mechaniczno-biologicznego przetwarzania zmieszanych odpadów komunalnych oraz pozostałości z sortowania odpadów komunalnych':
    return 'orange'
  elif row['Rodzaj instalacji'] == 'Instalacja Termicznego Przekształcania Odpadów Komunalnych - planowana':
    return 'darkred'
  return 'red'
```

```python
cities['color'] = cities.apply(select_marker_color, axis=1)
cities.head(3)
```

```python
my_map = folium.Map(
  location=[49.85790439, 19.9680245],
  zoom_starts=2
)
popup = folium.Popup(max_width=450,min_width=450)
for _, city in cities.iterrows():
  folium.Marker(
      location = [city['X'], city['Y']],
      popup = folium.Popup(f"Rodzaj instalacji:{city['Rodzaj instalacji']}\n_____________\n Nazwa i adres instalacji: {city['Nazwa i adres instalacji']}\n_____________\n Podmiot ekspoatujący intalację: {city['Podmiot ekspoatujący intalację']}\n_____________\n Maksymalne moce przerobowe [tys. Mg/rok]: {city['Maksymalne moce przerobowe [tys. Mg/rok]']}\n_____________\n Pojemność dyspozycyjn [tys/m3]: {city['Pojemność dyspozycyjn [tys/m3]']}\n_____________\n Status: {city['Status:']}\n_____________\n Maksymalne moce przerobowe części biologicznej: {city['Maksymalne moce przerobowe części biologicznej']}\n_____________\n Maksymalne moce przerobowe części mechanicznej: {city['Maksymalne moce przerobowe części mechanicznej']}"),
      #, min_width=100,max_width=400
      tooltip=city['Nazwa i adres instalacji'],#przy najechaniu
      icon = folium.Icon(color=city['color'], prefix='fa', icon="circle")
  ).add_to(my_map)
my_map
```

```python
my_map.save('odpady.html')
```
